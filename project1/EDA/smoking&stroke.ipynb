```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('stroke.csv')
data
```

```python
smoke_stroke = data[['smoking_status', 'stroke']]  
smoke_stroke
```

```python
unknown_data = smoke_stroke[smoke_stroke['smoking_status'] == 'Unknown'].index  # Unknow dataëŠ” ì‚­ì œ 

smoke_stroke = smoke_stroke.drop(unknown_data) 

```

```python
rows = ['never smoked', 'formerly smoked', 'smokes'] 
info = [[0, 0] for _ in range(len(rows))]
for i in smoke_stroke.index:
   
        
    idx = rows.index(smoke_stroke.loc[i]['smoking_status'])  

    if(smoke_stroke.loc[i]['stroke'] == 1):
        info[idx][0] += 1
    else:
        info[idx][1] += 1 
```

```python
info_no = []
info_yes = []

for i in range(3):
    info_yes.append(info[i][0])
    info_no.append(info[i][1])
```

```python
plt.ylabel('Patient number')
plt.bar(rows, info_no)
plt.bar(rows, info_yes) 
plt.legend(['No Stroke', 'Stroke'])

```
